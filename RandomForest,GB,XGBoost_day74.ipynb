```python
## Importing Libraries
from sklearn.ensemble import RandomForestClassifier
import numpy as np
```

```python
## Importing dataset
x=np.array([
    [1,3],
    [2,3],
    [3,5],
    [4,4],
    [5,6],
    [7,9],
    [8,10],
    [1,2],
    [3,4],
    [6,8]
])
print(len(x))
```

```python
y=np.array([0,0,1,1,1,1,1,0,1,1])
len(y)
```

```python
## Tuning Model
model = RandomForestClassifier(n_estimators=2,max_depth=2,max_features=2,random_state=42)
model.fit(x,y)
```

```python
## Predicting scores
pred = model.predict([[2,1]])
```

```python
pred
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6,6))
```

```python
## Plotting Tree
plot_tree(
    model.estimators_[1],
    feature_names=['Hours Studied','Marks Scored'],
    class_names=['Fail','Pass'],
    filled=True
)
plt.show()
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
x= np.array([
    [1,5],
    [2,6],
    [3,6],
    [4,5],
    [5,4],
    [6,4],
    [7,3],
    [8,10],
    [4,6],
    [5,7]
])
y=np.array([0,1,1,0,0,0,0,1,1,1])
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model = GradientBoostingClassifier(n_estimators=50,max_depth=2,learning_rate=0.2,random_state=42)
model.fit(x_train,y_train)
pred = model.predict(x_test)
print(x_test,pred)
```

```python
accuracy = accuracy_score(y_test,pred)
accuracy
```

```python
mod = XGBClassifier(n_estimators=50,max_depth=3,learning_rate=0.1,subsample=0.8,colsample_bytree=0.8,random_state=42)
mod.fit(x_train,y_train)
pr = model.predict(x_test)
print(x_test,pr)
```
